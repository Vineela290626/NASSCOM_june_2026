```python
# ===========================================================
# DATA CLEANING LAB - COMPLETE SOLUTION
# ===========================================================

import numpy as np
import pandas as pd

pd.set_option('display.max_columns', None)

# -----------------------------------------------------------
# MESSY DATASET
# -----------------------------------------------------------

raw = pd.DataFrame({
    'id':    [1, 2, 3, 4, 5, 6, 7, 7],
    'name':  ['Ana', 'Bo', 'Cy', 'Di', 'Eve', 'Fin', 'Gus', 'Gus'],
    'age':   [30, 25, np.nan, 41, -1, 38, 29, 29],
    'city':  [' Pune ', 'pune', 'DELHI', 'Delhi ', 'Mumbai', 'bombay', 'Pune.', 'Pune.'],
    'spend': ['120.5', '80.0', '200.2', 'N/A', '150.0', '99000', '110.0', '110.0'],
    'date':  ['2024-01-05', '2024-01-06', '2024-01-07', '2024-01-08',
              '2024-01-09', '2024-01-10', '2024-01-11', '2024-01-11'],
})

# ===========================================================
# LAB EXERCISE 1 - PROFILE DATA
# ===========================================================

print("========== LAB EXERCISE 1 ==========")

df = raw.copy()

print("Duplicate rows:", df.duplicated().sum())

print("\nMissing values per column:")
print(df.isna().sum())

# Problems:
# 1. Missing value in age
# 2. spend stored as text
# 3. spend contains 'N/A'
# 4. age contains -1 as disguised missing value
# 5. Duplicate row exists
# 6. Inconsistent city spellings
# 7. date stored as text

# ===========================================================
# LAB EXERCISE 2 - DROP VS IMPUTE
# ===========================================================

print("\n========== LAB EXERCISE 2 ==========")

ex = raw.copy()

# Unmask missing values
ex['spend'] = pd.to_numeric(ex['spend'], errors='coerce')
ex['age'] = ex['age'].replace(-1, np.nan)

# Drop version
drop_version = ex.dropna()

# Impute version
impute_version = ex.copy()
impute_version['age'] = impute_version['age'].fillna(
    impute_version['age'].median()
)
impute_version['spend'] = impute_version['spend'].fillna(
    impute_version['spend'].median()
)

print("Original rows :", len(ex))
print("Dropna rows   :", len(drop_version))
print("Imputed rows  :", len(impute_version))
print("Rows lost     :", len(ex) - len(drop_version))

# ===========================================================
# CLEAN MAIN DATAFRAME
# ===========================================================

df = raw.copy()

# Convert disguised missing values
df['spend'] = pd.to_numeric(df['spend'], errors='coerce')
df['age'] = df['age'].replace(-1, np.nan)

# Impute missing values
df['age'] = df['age'].fillna(df['age'].median())
df['spend'] = df['spend'].fillna(df['spend'].median())

# ===========================================================
# LAB EXERCISE 3 - DEDUPE & RETYPE
# ===========================================================

print("\n========== LAB EXERCISE 3 ==========")

ex = raw.copy()

ex['spend'] = pd.to_numeric(ex['spend'], errors='coerce')
ex['date'] = pd.to_datetime(ex['date'])

ex = ex.drop_duplicates()

print(ex.dtypes)
print("Shape:", ex.shape)

# -----------------------------------------------------------
# Continue cleaning main dataframe
# -----------------------------------------------------------

df = df.drop_duplicates()

df['date'] = pd.to_datetime(df['date'])
df['city'] = df['city'].astype('string')

# ===========================================================
# OUTLIER DETECTION
# ===========================================================

print("\n========== OUTLIER DETECTION ==========")

q1, q3 = df['spend'].quantile([0.25, 0.75])

iqr = q3 - q1

low = q1 - 1.5 * iqr
high = q3 + 1.5 * iqr

print("Q1:", q1)
print("Q3:", q3)
print("IQR:", iqr)

print("Lower Bound:", low)
print("Upper Bound:", high)

outliers = df[
    (df['spend'] < low) |
    (df['spend'] > high)
]

print("\nOutlier Rows:")
print(outliers[['name', 'spend']])

# Cap outliers
df['spend_capped'] = df['spend'].clip(
    lower=low,
    upper=high
)

# ===========================================================
# LAB EXERCISE 4 - AGE OUTLIERS
# ===========================================================

print("\n========== LAB EXERCISE 4 ==========")

q1_age = df['age'].quantile(0.25)
q3_age = df['age'].quantile(0.75)

iqr_age = q3_age - q1_age

low_age = q1_age - 1.5 * iqr_age
high_age = q3_age + 1.5 * iqr_age

print("Q1 Age:", q1_age)
print("Q3 Age:", q3_age)
print("IQR Age:", iqr_age)

print("Lower Bound:", low_age)
print("Upper Bound:", high_age)

age_outliers = df[
    (df['age'] < low_age) |
    (df['age'] > high_age)
]

print("\nAge Outliers:")
print(age_outliers[['name', 'age']])

# ===========================================================
# TEXT CLEANING
# ===========================================================

print("\n========== CITY BEFORE CLEANING ==========")
print(df['city'].value_counts())

s = df['city'].astype('string')

s = s.str.strip()
s = s.str.lower()
s = s.str.replace('.', '', regex=False)
s = s.replace({'bombay': 'mumbai'})

df['city'] = s

print("\n========== CITY AFTER CLEANING ==========")
print(df['city'].value_counts())

# ===========================================================
# LAB EXERCISE 5
# ===========================================================

print("\n========== LAB EXERCISE 5 ==========")

messy = pd.Series(
    [' London ', 'london', 'LONDON',
     'N.Y.', 'new york ', 'New York'],
    dtype='string'
)

messy = messy.str.strip().str.lower()

messy = messy.replace({
    'n.y.': 'new york'
})

print(messy.value_counts())

# ===========================================================
# FINAL CLEAN DATASET
# ===========================================================

clean = df.drop(columns=['spend_capped'])

print("\n========== FINAL DATASET ==========")

print("Final shape:", clean.shape)
print("Missing values:", int(clean.isna().sum().sum()))
print("Duplicates:", int(clean.duplicated().sum()))

print("\nClean Dataset:")
print(clean)
```
